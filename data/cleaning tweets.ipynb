```python
import pandas as pd
tweets = pd.read_csv('fb_sent.csv')
display(tweets)
dates = {}
```

```python
for i, row in tweets.iterrows():
    if row['date'][:7] not in dates:
        dates[row['date'][:7]] = {'ratings': row['rating'], 'count': 1}
    else:
        dates[row['date'][:7]]['ratings'] += row['rating']
        dates[row['date'][:7]]['count'] += 1
        
display(dates)
```

```python
scores = {}
for k, v in dates.items():
    scores[k] = v['ratings'] / v['count']
display(scores)
```

```python
pd.DataFrame.from_dict(scores, orient='index', columns=['rating']).to_csv('fb_clean.csv')
```
